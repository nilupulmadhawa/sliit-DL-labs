```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

#1
array_4x4 = np.random.exponential(scale=1, size=(4, 4))

print("Random 4x4 array from exponential distribution:")
print(array_4x4)

print('\n')

#2
array_100000x1 = np.random.exponential(scale=1, size=100000)

plt.hist(array_100000x1, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="orange", label="Uniform")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="green", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

print('\n')

#3

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title("3D Plot of Z = X^2 + Y^2")
plt.show()

print('\n')


```
